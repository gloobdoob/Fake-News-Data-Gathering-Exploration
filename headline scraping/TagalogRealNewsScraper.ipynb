```python
from requests_html import HTMLSession
from collections import defaultdict
session = HTMLSession()
```

```python
real_dic = defaultdict(list)
```

```python
abs_headlines = []
for n in range(1, 12):
    url_abs = f'https://news.abs-cbn.com/patrol/tag/tagalog-news?page={n}'
    r = session.get(url_abs)
    articles = r.html.find('article')
    for item in articles:
        newsitem = item.find('a')
        abs_headlines.append(newsitem[1].text)
        print(newsitem[1].text)
```

```python
len(abs_headlines)
```

```python
gma_headlines = []
for n in range(117, 122):
    r = session.get(f'https://data.gmanews.tv/gno/widgets/grid_reverse_listing/story_btbbalita/{n}')
    gma_data = r.json()
    for article in gma_data['data']:
        gma_headlines.append(article['title'])
        print(article['title'])
```

```python
gma_headlines = list(dict.fromkeys(gma_headlines))
print(gma_headlines)
len(gma_headlines)
```

```python
from bs4 import BeautifulSoup
import requests

phs_headlines = []
for n in range(0, 40):
    r = session.get(f'https://www.philstar.com/micro_page.php?qcount=1&micro=77&page={n}&cell=2&lastid=2260891')
    soup = BeautifulSoup(r.content, "html.parser")
    articles = soup.find_all("div", {"class":"microsite_article_title"})
    for headline in articles:
        phs_headlines.append(headline.text)
        print(headline.text)

            


    
```

```python
list_of_dates = []
yrs = [2023, 2022, 2021]
for yr in yrs:
    for month in range(12, 0, -1):
        for day in range(31, 0, -1):
            month = str(month)
            day = str(day)
            if len(month) < 2:
                month = '0'+ month
            if len(day) < 2:
                day = '0'+ day
            list_of_dates.append(f'{yr}-{month}-{day}')

```

```python
list_of_dates = list_of_dates[260:]
```

```python
ds_headlines = []
#for n in range(1,30):

for date in list_of_dates:
    r = session.get(f'https://www.thedailysentry.net/search?updated-max={date}T08%3A49%3A00-08%3A00&max-results=8')
    soup = BeautifulSoup(r.content, "html.parser")
    articles = soup.find_all("h2", {"class":"post-title entry-title"})
    for headline in articles:
        hl = headline.text.replace('\n', '')
        if hl not in ds_headlines:
            ds_headlines.append(hl)
            print(hl)
        else:
            continue
    print(date)
```
