```python
abs_headlines = []
for n in range(1, 12):
    url_abs = f'https://news.abs-cbn.com/patrol/tag/tagalog-news?page={n}'
    r = session.get(url_abs)
    articles = r.html.find('article')
    for item in articles:
        newsitem = item.find('a')
        abs_headlines.append(newsitem[1].text)
        print(newsitem[1].text)
```

```python
list_of_dates = []
yrs = [2023, 2022, 2021]
for yr in yrs:
    for month in range(12, 0, -1):
        for day in range(31, 0, -1):
            month = str(month)
            day = str(day)
            if len(month) < 2:
                month = '0'+ month
            if len(day) < 2:
                day = '0'+ day
            list_of_dates.append(f'{yr}-{month}-{day}')

```

```python
list_of_dates = list_of_dates[260:]
```

```python
ds_headlines = []
#for n in range(1,30):

for date in list_of_dates:
    r = session.get(f'https://www.thedailysentry.net/search?updated-max={date}T08%3A49%3A00-08%3A00&max-results=8')
    soup = BeautifulSoup(r.content, "html.parser")
    articles = soup.find_all("h2", {"class":"post-title entry-title"})
    for headline in articles:
        hl = headline.text.replace('\n', '')
        if hl not in ds_headlines:
            ds_headlines.append(hl)
            print(hl)
        else:
            continue
    print(date)
```

```python

from bs4 import BeautifulSoup
def replace_end(article, rep_dic):
    #rep_dic = {', article with image':'', ', article with video':'', ', article with gallery': '' }
    for i, j in rep_dic.items():
        article = article.replace(i, j)
    return article

def fake_scraper(url, element, n_pages, site_abbv, rep_dic=None ,test=False):
    headlines = []
    for n in range(1, n_pages):
        r = session.get(url.format(page=n))
        soup = BeautifulSoup(r.content, "html.parser")
        articles = soup.find_all(element)
        for headline in articles:
            acc_text = headline.text
            if acc_text != '':

                if rep_dic is not None:
                    acc_text = replace_end(acc_text, rep_dic)

                headlines.append(acc_text)
                print(acc_text)
        if test == True:
            break
    
    if test == True:
        return headlines
    
    print(len(headlines))
    file = open(f'{site_abbv}_news.txt','w')
    for news in headlines:
        file.write(news+"\n")
    file.close()
    return headlines
```

```python

bgd_headlines = fake_scraper('https://brigadanews.ph/category/national/page/{page}/', 
                             ('h2', {'class':"entry-title"}), 
                             50, 
                             'bgd', 
                             rep_dic=None, 
                             test=False)
```
