```python
import pandas as pd

df = pd.read_csv('VeraFilesDataset.csv',delimiter=',')
df = df.fillna("")
df = df.applymap(lambda x: x.replace('"', ''))
df
```

```python
from lingua import Language, LanguageDetectorBuilder
languages = [Language.ENGLISH, Language.TAGALOG]
detector = LanguageDetectorBuilder.from_languages(*languages).build()
```

```python
df['Statement_Filipino'].iloc[154]
```

```python
str(detector.detect_language_of(df['Statement_Filipino'].iloc[154])).replace('Language.', '')
```

```python
def lang_detector(x):
    return str(detector.detect_language_of(x)).replace('Language.', '')

df['Statement_Filipino_Language'] = df['Statement_Filipino'].apply(lambda x : lang_detector(x))
df['Statement_English_Language'] = df['Statement_English'].apply(lambda x : lang_detector(x))


```

```python
import numpy as np
df['Statement_English'] = np.where((df['Statement_Filipino_Language'] == 'ENGLISH'), df['Statement_Filipino'], df['Statement_English'])
```

```python
df['Statement_Filipino'] = np.where((df['Statement_Filipino_Language'] == 'ENGLISH'), '', df['Statement_Filipino'])
```

```python
df.to_csv('VeraFilesDataset1.csv', encoding='utf-8-sig', index=False)
```
