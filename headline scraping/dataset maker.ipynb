```python
with open('text files/real_eng_news.txt') as f:
    lines = f.readlines()
    real_eng_news = [line.replace('\n','') for line in lines]

with open('text files/fake_eng_news.txt') as f:
    lines = f.readlines()
    fake_eng_news = [line.replace('\n','') for line in lines]
```

```python
real_eng_news
```

```python
import random
import pandas as pd
```

```python
random.shuffle(real_eng_news)
random.shuffle(fake_eng_news)
```

```python
len(real_eng_news)
```

```python
len(fake_eng_news)
```

```python
fake_eng_news = fake_eng_news[:len(real_eng_news)]
```

```python
from collections import defaultdict
headlines = defaultdict(list)

for news in real_eng_news:
    headlines['Statement'].append(news)
    headlines['Rating'].append('Real')

for news in fake_eng_news:
    headlines['Statement'].append(news)
    headlines['Rating'].append('Fake')
```

```python
df = pd.DataFrame(headlines)
```

```python
df.to_csv('Scraped_English_Headlines.csv', index=False, encoding='utf-8-sig')
```

```python
with open('real_fil_news.txt') as f:
    lines = f.readlines()
    real_fil_news = [line.replace('\n','') for line in lines]

with open('fake_fil_news.txt') as f:
    lines = f.readlines()
    fake_fil_news = [line.replace('\n','') for line in lines]
```

```python
vfiles_df = pd.read_csv('VeraFilesDataset.csv')
vfiles_df = vfiles_df[['Statement_Filipino', 'Rating']]
vfiles_df
```

```python
fakements = vfiles_df[vfiles_df['Rating']=='Fake']['Statement_Filipino'].values
```

```python
fakements = list(fakements)
fake_fil_news = fake_fil_news + fakements
```

```python
random.shuffle(real_fil_news)
random.shuffle(fake_fil_news)
```

```python
len(real_fil_news)
```

```python
len(fake_fil_news)
```

```python
real_fil_news = real_fil_news[:len(fake_fil_news)]
```

```python
from collections import defaultdict
fil_headlines = defaultdict(list)

for news in real_fil_news:
    fil_headlines['Statement'].append(news)
    fil_headlines['Rating'].append('Real')

for news in fake_fil_news:
    fil_headlines['Statement'].append(news)
    fil_headlines['Rating'].append('Fake')
```

```python
df_fil = pd.DataFrame(fil_headlines)
```
